```python
from astropy.table import Table
import numpy as np
```

```python
#import the SGA catalog data
SGA = Table.read('/global/cfs/cdirs/cosmo/data/sga/2020/SGA-2020.fits',hdu=1)
trac = Table.read('/global/cfs/cdirs/cosmo/data/sga/2020/SGA-2020.fits',hdu=2)
```

```python
#create columns to store needed tractor data
SGA.add_column(np.nan, name = 'FLUX_G')
SGA.add_column(np.nan, name = 'FLUX_R')
SGA.add_column(np.nan, name = 'FLUX_Z')
SGA.add_column(np.nan, name = 'PSFDEPTH_G')
SGA.add_column(np.nan, name = 'PSFDEPTH_R')
SGA.add_column(np.nan, name = 'PSFDEPTH_Z')
SGA.add_column(np.nan, name = 'PSFSIZE_G')
SGA.add_column(np.nan, name = 'PSFSIZE_R')
SGA.add_column(np.nan, name = 'PSFSIZE_Z')

SGA.add_column(np.nan, name='RELEASE')
SGA.add_column(np.nan, name='BRICKID')
SGA.add_column(np.nan, name='OBJID')
SGA.add_column(np.nan, name='MASKBITS')
SGA.add_column(np.nan, name='SOURCE_TYPE')
SGA.add_column(np.nan, name='EBV')
SGA.add_column(np.nan, name='FIBERFLUX_G')
SGA.add_column(np.nan, name='FIBERFLUX_R')
SGA.add_column(np.nan, name='FIBERFLUX_Z')
SGA.add_column(np.nan, name='NOBS_G')
SGA.add_column(np.nan, name='NOBS_R')
SGA.add_column(np.nan, name='NOBS_Z')
SGA.add_column(np.nan, name='FRACIN_G')
SGA.add_column(np.nan, name='FRACIN_R')
SGA.add_column(np.nan, name='FRACIN_Z')
```

```python
#import tractor data
SGA['FLUX_G'] = trac['FLUX_G']
SGA['FLUX_R'] = trac['FLUX_R']
SGA['FLUX_Z'] = trac['FLUX_Z']
SGA['PSFDEPTH_G'] = trac['PSFDEPTH_G']
SGA['PSFDEPTH_R'] = trac['PSFDEPTH_R']
SGA['PSFDEPTH_Z'] = trac['PSFDEPTH_Z']
SGA['PSFSIZE_G'] = trac['PSFSIZE_G']
SGA['PSFSIZE_R'] = trac['PSFSIZE_R']
SGA['PSFSIZE_Z'] = trac['PSFSIZE_Z']
SGA['RELEASE'] = trac['RELEASE']
SGA['BRICKID'] = trac['BRICKID']
SGA['OBJID'] = trac['OBJID']
SGA['MASKBITS'] = trac['MASKBITS']
SGA['SOURCE_TYPE'] = trac['TYPE']
SGA['EBV'] = trac['EBV']
SGA['FIBERFLUX_G'] = trac['FIBERFLUX_G']
SGA['FIBERFLUX_R'] = trac['FIBERFLUX_R']
SGA['FIBERFLUX_Z'] = trac['FIBERFLUX_Z']
SGA['NOBS_G'] = trac['NOBS_G']
SGA['NOBS_R'] = trac['NOBS_R']
SGA['NOBS_Z'] = trac['NOBS_Z']
SGA['FRACIN_G'] = trac['FRACIN_G']
SGA['FRACIN_R'] = trac['FRACIN_R']
SGA['FRACIN_Z'] = trac['FRACIN_Z']

### THIS DOES NOT HAVE NEEDED VALUES OF: EXTINCTION, PHOTOMETRIC REDSHIFT MEDIAN
```
